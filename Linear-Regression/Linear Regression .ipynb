### Compute the Cost $J(\omega)$
---

The objective of linear regression is to minimize the cost function

$$J(\omega) = \frac{1}{m} \sum_{i=1}^m (h_\omega(x_i) - y_i )^2$$

where $h_{\omega}(x)$ is the hypothesis and given by the linear model

$$h_{\omega}(x) = \omega_0 + \omega_1x$$

```python
def cost_function(X, y, omega):
    m = len(y)
    y_pred = X.dot(omega)
    error = (y_pred - y) ** 2
    
    return 1 / m * np.sum(error)
```

```python
m = data.Population.values.size
X = np.append(np.ones((m, 1)), data.Population.values.reshape(m, 1), axis=1)
y = data.Profit.values.reshape(m, 1)
omega = np.zeros((2,1))

cost_function(X, y, omega)
```

### Gradient Descent
---

Minimize the cost function $J(\omega)$ by updating the below equation and repeat unitil convergence
        
$$\omega_0 := \omega_0 - \alpha \frac{2}{m} \sum_{i=1}^m (h_{\omega}(x_i) - y_i)$$

$$\omega_1 := \omega_1 - \alpha \frac{2}{m} \sum_{i=1}^m (h_{\omega}(x_i) - y_i)x_i$$

```python
def gradient_descent(X, y, omega, alpha, epochs): 
    m = len(y)
    costs = []
    for i in range(epochs):
        y_pred = X.dot(omega)
        error = np.dot(X.transpose(), (y_pred - y))
        omega -= alpha * 2/m * error
        costs.append(cost_function(X, y, omega))
        
    return omega, costs
```

```python
omega, costs = gradient_descent(X, y, omega, alpha=0.01, epochs=1000)

print("h(x) = {} + {}x".format(str(round(omega[0, 0], 2)),
                                str(round(omega[1, 0], 2))))
```

### Visualising the Cost Function $J(\omega)$
---

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
omega_0 = np.linspace(-10,10,100)
omega_1 = np.linspace(-5,5,100)

cost_values = np.zeros((len(omega_0), len(omega_1)))

for i in range(len(omega_0)):
    for j in range(len(omega_1)):
        w = np.array([omega_0[i], omega_1[j]])
        cost_values[i, j] = cost_function(X, y, w)
```

```python
fig = plt.figure(figsize = (12, 8))
ax = fig.gca(projection = '3d')

surf = ax.plot_surface(omega_0, omega_1, cost_values, cmap = "viridis", linewidth = 0.2)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.xlabel("$\omega_0$")
plt.ylabel("$\omega_1$")
ax.set_zlabel("$J(\omega)$")
ax.set_title("Cost Surface")
ax.view_init(30,330)

plt.show()
```

### Predicting the profit using $\omega$ values
---

$h_{\omega}(x) = \omega_0 + \omega_1x$

```python
def predict(x, omega):
    y_pred = np.dot(omega.transpose(), x)
    return y_pred
```

```python
y_pred = predict(np.array([1, 4]),omega) * 10000
print("For a population of 40,000, the model predicts a profit of $" + str(round(y_pred, 0)))
```
